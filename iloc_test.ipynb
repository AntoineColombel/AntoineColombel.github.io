```python
!pip install datasets matplotlib

from datasets import load_dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_DA_FR = df[(df['job_title_short'].isin(job_titles)) & (df['job_country'] == 'France')].copy()

df_DA_FR = df_DA_FR.dropna(subset=['salary_year_avg'])

job_list = [df_DA_FR[df_DA_FR['job_title_short']== job_title]['salary_year_avg'] for job_title in job_titles]

plt.boxplot(job_list, labels=job_titles, vert=False)
plt.show()
```
